```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('owid-covid-data.csv')
df['date'] = pd.to_datetime(df['date'])

countries = ['United States', 'India', 'Brazil', 'Russia', 'United Kingdom']

df_countries = df[df['location'].isin(countries)]

# Plot a line chart of total cases over time
plt.figure(figsize=(12, 6))

for country in countries:
    country_data = df_countries[df_countries['location'] == country]
    plt.plot(country_data['date'], country_data['total_cases'], label=country)

# Chart formatting
plt.title('Total COVID-19 Cases Over Time')
plt.xlabel('Date')
plt.ylabel('Total Cases')
plt.legend()
plt.grid(True)
plt.tight_layout()

# Show the chart
plt.show()
```

```python
# Get the latest date's data
latest_date = df['date'].max()
latest_df = df[df['date'] == latest_date]

# Top 10 countries by total cases
top_countries = latest_df.sort_values('total_cases', ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(data=top_countries, x='location', y='total_cases', palette='viridis')
plt.xticks(rotation=45)
plt.title('Top 10 Countries by Total Cases')
plt.xlabel('Country')
plt.ylabel('Total Cases')
plt.show()
```

```python

features = ['total_cases', 'total_deaths', 'new_cases', 'new_deaths', 'people_vaccinated']
corr_df = df[features].dropna()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of COVID-19 Metrics')
plt.tight_layout()
plt.show()
```

```python

countries = ['United States', 'India', 'Brazil', 'Russia', 'United Kingdom']
df_vax = df[df['location'].isin(countries)]
```

```python
plt.figure(figsize=(12, 6))

for country in countries:
    country_data = df_vax[df_vax['location'] == country]
    plt.plot(country_data['date'], country_data['people_vaccinated'], label=country)

plt.title('Cumulative Vaccinations Over Time')
plt.xlabel('Date')
plt.ylabel('People Vaccinated')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python

latest_vax = df_vax[df_vax['date'] == df_vax['date'].max()]
latest_vax = latest_vax[['location', 'people_vaccinated_per_hundred']].dropna()
plt.figure(figsize=(10, 6))
sns.barplot(x='people_vaccinated_per_hundred', y='location', data=latest_vax)
plt.title('Vaccinated Population (% of Total) - Latest Available Data')
plt.xlabel('% of Population Vaccinated')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```

```python
Key Insights from COVID-19 & Vaccination Data
1. 🇺🇸 The United States had the fastest early vaccine rollout.
The U.S. led in cumulative vaccinations during the initial stages, with a sharp increase in early 2021. This was evident in the cumulative vaccination line chart, where the U.S. curve rose faster than most other countries.

2. 🇮🇳 India experienced the steepest growth in total cases.
India showed a sharp rise in total cases, especially during the second wave in 2021. The line chart of total cases over time highlighted a major spike compared to other countries.

3. ⚖️ Death rates vary significantly across countries.
Some countries (like Brazil and Russia) exhibited a higher death rate (total deaths / total cases) compared to others, despite similar case counts. This may reflect differences in healthcare systems, testing, and reporting.

4. 🌍 Vaccination coverage is uneven.
While countries like the UK and US have over 70% of their populations vaccinated, others like Russia lag behind. This is clearly shown in the % vaccinated population bar chart.

5. ⚠️ Anomalies in reporting and missing data
Certain countries have visible gaps or flat lines in vaccination data, suggesting delays or inconsistencies in reporting. These anomalies are apparent in line charts where data points stop abruptly or plateau unnaturally.
```
